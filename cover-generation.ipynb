```python
from openai import OpenAI
import toml

# read api_key from env/secrets.toml
secrets = toml.load("env/secrets.toml")
api_key = secrets["openai"]["api_key"]


prompt = "Representa la valentia, la sabidur√≠a y la maldad en tres personajes diferentes. El primero es un guerrero valiente, el segundo es un sabio anciano y el tercero es un malvado brujo. Representalos interactuando juntos."
client = OpenAI(api_key=api_key)

response = client.images.generate(
  model="dall-e-3",
  prompt=prompt,
  size='1792x1024',
  quality="standard",
  n=1,
)

image_url = response.data[0].url

```

```python
# display image
import requests

img_data = requests.get(image_url).content

import matplotlib.pyplot as plt

with open('image-2.jpg', 'wb') as handler:
    handler.write(img_data)

img = plt.imread('image-2.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
```
